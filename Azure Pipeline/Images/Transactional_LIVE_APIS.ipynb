```python
def sanitize_table_name(name):
    return re.sub(r'\W+', '_', name)

def fetch_paginated_data(url, headers):
    all_data = []
    retries = 0
    while url:
        try:
            response = requests.get(url, headers=headers)
            if response.status_code == 204:
                return []  # No content
            data = response.json()
            all_data.extend(data.get("value", []))
            url = data.get("@odata.nextLink", None)
        except Exception as e:
            if retries < 3:
                time.sleep(2)
                retries += 1
                continue
            raise e
    return all_data
```

```python
for raw_name, url in tables:
    table_name = sanitize_table_name(raw_name)
    print(f"ðŸ”„ Fetching data for: {table_name}")
    try:
        records = fetch_paginated_data(url, headers)

        if not records:  # If empty, skip without dropping
            print(f"âš ï¸ No data found in {table_name}, keeping existing table")
            continue

        # Convert to DataFrame
        df = pd.DataFrame(records)
        spark_df = spark.createDataFrame(df)

        # Drop only when new data exists
        spark.sql(f"DROP TABLE IF EXISTS dbo.{table_name}")

        # Save the new table inside dbo schema
        spark_df.write.mode("overwrite").saveAsTable(f"dbo.{table_name}")

        print(f"âœ… Saved: {table_name} ({len(records)} rows)")

    except Exception as e:
        print(f"âŒ Error fetching {table_name}: {str(e)}")
```
